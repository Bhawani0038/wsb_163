```python
import matplotlib.pyplot as plt
import numpy as np

# Dates for the week
dates = np.arange(np.datetime64('2025-03-17'), np.datetime64('2025-03-24'))

# Maximum temperatures for each day (in C)
max_temps = [33.4, 34.0, 34.5, 35.0, 35.5, 36.0, 36.5]

# Minimum temperatures for each day (in C)
min_temps = [18.0, 18.5, 19.0, 19.5, 20.0, 20.5, 21.0]

# Plotting the bar chart
bar_width = 0.4
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting max temperatures
ax.bar(dates - np.timedelta64(12, 'h'), max_temps, width=bar_width, label='Max Temp (C)', color='tomato')

# Plotting min temperatures
ax.bar(dates + np.timedelta64(12, 'h'), min_temps, width=bar_width, label='Min Temp (C)', color='royalblue')

# Formatting the plot
ax.set_xlabel('Date')
ax.set_ylabel('Temperature (C)')
ax.set_title('Daily Maximum and Minimum Temperatures in Jodhpur (March 1723, 2025)')
ax.set_xticks(dates)
ax.set_xticklabels([date.astype(str) for date in dates], rotation=45)
ax.legend()

# Display the plot
plt.tight_layout()
plt.show()
```

```python
import os
import subprocess
import webbrowser
import time
import psutil

def open_notepad():
    subprocess.Popen("notepad.exe")
    print("Ultron: Notepad opened.")

def close_notepad():
    for process in psutil.process_iter(attrs=['pid', 'name']):
        if process.info['name'] == 'notepad.exe':
            os.kill(process.info['pid'], 9)
            print("Ultron: Notepad closed.")
            return
    print("Ultron: Notepad is not running.")

def open_youtube():
    webbrowser.open("https://www.youtube.com")
    print("Ultron: YouTube opened.")

def close_browser():
    for process in psutil.process_iter(attrs=['pid', 'name']):
        if 'chrome' in process.info['name'].lower() or 'msedge' in process.info['name'].lower() or 'firefox' in process.info['name'].lower():
            os.kill(process.info['pid'], 9)
            print("Ultron: Browser closed.")
            return
    print("Ultron: No browser found.")

def main():
    while True:
        command = input("\nUltron listening... Enter your command: ").lower()

        if "open notepad" in command:
            open_notepad()
        elif "close notepad" in command:
            close_notepad()
        elif "open youtube" in command:
            open_youtube()
        elif "close youtube" in command or "close browser" in command:
            close_browser()
        elif "exit" in command:
            print("Ultron: Shutting down...")
            break
        else:
            print("Ultron: Sorry, I didn't understand that command.")

if __name__ == "__main__":
    main()
```

```python
import os
import subprocess
import webbrowser
import psutil
import speech_recognition as sr
import pyttsx3

# Initialize text-to-speech engine
engine = pyttsx3.init()
engine.setProperty('rate', 150)  # Speed of speech
engine.setProperty('volume', 1.0)  # Volume

def speak(text):
    """Speak out the given text."""
    engine.say(text)
    engine.runAndWait()

def listen():
    """Listen to user commands via microphone."""
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Ultron is listening...")
        recognizer.adjust_for_ambient_noise(source)
        try:
            audio = recognizer.listen(source, timeout=5)
            command = recognizer.recognize_google(audio).lower()
            print(f"You said: {command}")
            return command
        except sr.UnknownValueError:
            print("Ultron: Sorry, I couldn't understand.")
            return None
        except sr.RequestError:
            print("Ultron: Error connecting to speech service.")
            return None
        except sr.WaitTimeoutError:
            print("Ultron: No voice detected.")
            return None

def open_notepad():
    subprocess.Popen("notepad.exe")
    speak("Opening Notepad")
    print("Ultron: Notepad opened.")

def close_notepad():
    for process in psutil.process_iter(attrs=['pid', 'name']):
        if process.info['name'] == 'notepad.exe':
            os.kill(process.info['pid'], 9)
            speak("Closing Notepad")
            print("Ultron: Notepad closed.")
            return
    speak("Notepad is not running.")
    print("Ultron: Notepad is not running.")

def open_youtube():
    webbrowser.open("https://www.youtube.com")
    speak("Opening YouTube")
    print("Ultron: YouTube opened.")

def close_browser():
    for process in psutil.process_iter(attrs=['pid', 'name']):
        if 'chrome' in process.info['name'].lower() or 'msedge' in process.info['name'].lower() or 'firefox' in process.info['name'].lower():
            os.kill(process.info['pid'], 9)
            speak("Closing browser")
            print("Ultron: Browser closed.")
            return
    speak("No browser found.")
    print("Ultron: No browser found.")

def main():
    speak("Hello! I am Ultron. How can I assist you?")
    while True:
        command = listen()
        if command:
            if "open notepad" in command:
                open_notepad()
            elif "close notepad" in command:
                close_notepad()
            elif "open youtube" in command:
                open_youtube()
            elif "close youtube" in command or "close browser" in command:
                close_browser()
            elif "exit" in command or "shut down" in command:
                speak("Goodbye! Have a great day!")
                print("Ultron: Shutting down...")
                break
            else:
                speak("Sorry, I didn't understand that command.")
                print("Ultron: Sorry, I didn't understand that command.")

if __name__ == "__main__":
    main()
```
