```python
# Initialize an empty cart

cart = {}     #{"laptop" : [2, 50000], "keyboard" : [1, 1000]}
inventory = {
    "Laptop"  : {"price" : 50000, "stock" : 20}, 
    "Speaker" : {"price" : 5000, "stock" : 30},
    "Keyboard": {"price" : 1000, "stock" : 10}
}

def display_inventory():  #available product in store
    print("Product     Price      Stock")
    for product_name, v in inventory.items():
        print(f"{product_name: <10}|{v['price'] : <10}|{v['stock']}")

def add_to_cart():   #your personal cart
    display_inventory()
    p_name = input("what to purchase:")   #keyboard
    stock = inventory.get(p_name).get("stock")
    if p_name in inventory and stock > 0:
        price = inventory.get(p_name).get("price")
        quantity = int(input("How many quantity to purchase:"))
        cart[p_name] =  [quantity, price]      

def remove_from_cart():
    view_cart()
    p_name = input("Enter product name to remove:")
    if p_name in cart:
        cart.pop(p_name)
    else:
        print("No such element exist")

def view_cart():
    print("Product     Price      Quantity")
    for product_name, v in cart.items():
        print(f"{product_name: <10}|{v[1] : <10}|{v[0]}")


def generate_invoice():   #bill
    total_bill = 0
    print("Product     Price      Quantity")
    for product_name, v in cart.items():
        print(f"{product_name: <10}|{v[1] : <10}|{v[0]}")
        total_bill =  total_bill + v[1] * v[0]
        inventory[product_name]["stock"] = inventory[product_name]["stock"] - v[0]
    print("Total Bill = ", total_bill)

def main():
    while True:
        print("\n=== Inventory Management System ===")
        print("1. View Inventory")
        print("2. Add Item to Cart")
        print("3. Remove Item from Cart")
        print("4. View Cart")
        print("5. Generate Invoice")
        print("6. Exit")
        choice = input("Enter your choice (1-6): ").strip()
        if choice == '1':
            display_inventory()
        elif choice == '2':
            add_to_cart()
        elif choice == '3':
            remove_from_cart()
        elif choice == '4':
            view_cart()
        elif choice == '5':
            generate_invoice()
        elif choice == '6':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("❌ Invalid choice. Please select a valid option.")

main()
```
