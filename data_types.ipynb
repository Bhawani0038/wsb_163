# 1. What is a Variable?
A variable is a name that is used to store data in a program. 
<br>It acts as a container that holds values, which can be changed during the execution of the program.

## 1.1 Characteristics of Variables:
    - Stores data – Holds values like numbers, text, etc.
    - Has a unique name – Used to reference the stored data.
    - Value can change – Can be updated with new values.
    - Memory allocation – Reserves space in memory for storage.

## 1.2 Declaring Variables in Python
In Python, you can create a variable by simply assigning a value to a name using the = operator.

```python
# Assigning values to variables
name = "Wscube"    
age = 25          
height = 5.6     
is_student = True  

# Printing the variables
print(name)      
print(age)        
print(height)    
print(is_student) 
```

# 3. What are Data Types?

A data type defines the kind of data a variable can store in a programming language. <br>It helps the computer understand how to process and use the data efficiently.

## 3.1 Primary Data (Primittive Data Type):
A primary data type (also called a primitive data type) is the most basic type of data that a programming language provides.<br>
These are the building blocks for storing and handling data in a program.

## 3.2 What is Derived Data / Collection Data Type in Python?
Derived data refers to data that is created from existing data using computations, transformations, or manipulations. 
<br>It is not directly stored but is generated by applying operations on primary data types.
