# 5. Membership Operator

Membership operators in and not in are used to check whether a value exists in a sequence (like a list, tuple, string, or dictionary).

## 1. Checking if an Element Exists in a List

```python
fruits = ["apple", "banana", "mango"]
print("apple" in fruits)   # True
print("grape" in fruits)    # False
```

## 2. Checking if a Character Exists in a String

```python
word = "Python"
print("P" in word)      # True
print("z" in word)      # False
print("Py" in word)     # True
```

## 3. Checking for a Key in a Dictionary

```python
student = {"name": "John", "age": 20, "grade": "A"}
print("name" in student)      # True (checks keys only)
print("John" in student)      # False (values are NOT checked)
print("grade" not in student) # False ("grade" exists)
```

## 4. Using not in for Membership Check

```python
sentence = "Welcome to Python!"
print("Java" not in sentence)   # True
print("Python" not in sentence) # False
```

# 6. Identity Operator

In Python, identity operators are used to compare memory locations of two objects. They check whether two variables refer to the same object in memory.

## 1.  Checking Identity with is

```python
a = [1, 2, 3]
b = a  # Both variables point to the same object
c = [1, 2, 3]  # New object with same values

print(a is b)   # True (same memory location)
print(a is c)   # False (different memory location)
print(a == c)   # True (same values but different objects)
```

## 2.  Checking Identity with is not

```python
x = 10
y = 10
z = 20

print(x is not y)  # False (same memory location for small integers)
print(x is not z)  # True (different objects)
```

## Difference Between is and ==

```python
a = [1, 2, 3]
b = [1, 2, 3]

print(a is b)   # False (different objects)
print(a == b)   # True (same values)
```
